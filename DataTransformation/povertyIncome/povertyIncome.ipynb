## Data Transformation: Table Poverty-Income

### Datasets used:
- Poverty data 2020
- US minimum wage by state
- US unemployment data

```python
import pandas as pd
```

Load data to notebook

```python
#Load data with pandas
poverty = pd.read_excel("est20us.xls", header=[2,3])
wage = pd.read_csv("Minimum_Wage_Data.csv", encoding='cp1252')
unemployment = pd.read_csv("unemployment_rate_data.csv", encoding='cp1252')
```

Cleaning data

```python
#Cleaning Poverty dataFrame

#Drop columns
poverty = poverty.drop(labels=["Age 0-17", "Age 5-17 in Families", "Age 0-4"], axis=1, level=0)
poverty = poverty.droplevel(level=0, axis=1)
poverty = poverty.drop(labels=(["State FIPS Code", "Postal Code", "90% CI Lower Bound", "90% CI Upper Bound", "90% CI Lower Bound.1", "90% CI Upper Bound.1", "90% CI Lower Bound", "90% CI Upper Bound"]), axis=1)

#Rename columns
poverty.columns = ['state', 'povertyEstimate', 'povertyPercent', 'medianHouseholdIncome']

#Data available just for 2020
poverty.head(1)
```

```python
#Cleaning Minimum wage dataFrame

#Keeping data just for 2020
wage = (wage[wage.Year == 2020]).reset_index()

#Drop columns
wage = wage.drop(labels=(["index", "Year", "State.Minimum.Wage", "Federal.Minimum.Wage",	"Federal.Minimum.Wage.2020.Dollars", "Effective.Minimum.Wage", "Effective.Minimum.Wage.2020.Dollars", "CPI.Average", "Department.Of.Labor.Uncleaned.Data",	"Department.Of.Labor.Cleaned.Low.Value", "Department.Of.Labor.Cleaned.Low.Value.2020.Dollars",	"Department.Of.Labor.Cleaned.High.Value", "Department.Of.Labor.Cleaned.High.Value.2020.Dollars", "Footnote"]), axis=1)

#Rename columns
wage.columns = ['state', 'minimumWage2020']

#Data available just for 2020
wage.head(1)
```
