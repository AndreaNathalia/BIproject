```python
import pandas as pd
```

```python
covid = pd.read_csv("us_counties_covid19_daily.csv")
cases = pd.read_csv("datacommons_data.csv")
deaths = pd.read_csv("datacommons_deaths.csv")
```

```python
covid = covid.drop(labels=["fips"], axis=1)
covid["date"] = pd.to_datetime(covid["date"])
lastDate = covid["date"].max()
lastReport2020 = covid.loc[covid['date'] >= lastDate]
lastReport2020 = lastReport2020.groupby("state").sum()
lastReport2020
```

```python
cases = cases.drop(labels=["placeDcid"], axis=1)
cases = cases.rename(columns={"placeName":"state", "CumulativeCount_MedicalConditionIncident_COVID_19_ConfirmedOrProbableCase":"positiveCases"})
deaths = deaths.drop(labels=["placeDcid"], axis=1)
deaths = deaths.rename(columns={"placeName":"state", "CumulativeCount_MedicalConditionIncident_COVID_19_PatientDeceased":"deaths"})
casesAll = cases.join(deaths.set_index("state"), on="state")
casesAll
```

Generate csv output

```python
casesAll.to_csv("covid.csv", sep=',')
```
